```python
import pandas as pd
import numpy as np
from sklearn.model_selection import RandomizedSearchCV,train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from xgboost import XGBClassifier
```

```python
df = pd.read_csv('train.csv')
dfcopy = df.copy()
```

```python
dfcopy.head()
```

```python
dfcopy.drop(columns=['id','Name', 'CGPA','Degree', 'Profession'], axis=1, inplace=True)
```

```python
dfcopy.head()
```

```python
dfcopy.drop(columns='City', axis=1, inplace= True)
```

```python
dfcopy.isnull().sum()
```

```python
dfcopy['Depression'].drop_duplicates(inplace=True)
```

```python
df['Sleep Duration'].value_counts()
```

```python
dfcopy['Financial Stress'].fillna(dfcopy['Financial Stress'].mean(),inplace=True)
```

```python
dfcopy['Study Satisfaction'].fillna(0,inplace=True)
dfcopy['Job Satisfaction'].fillna(0,inplace=True)
dfcopy['Academic Pressure'].fillna(0, inplace=True)
dfcopy['Work Pressure'].fillna(0, inplace=True)
```

```python
dfcopy['Pressure'] = dfcopy['Academic Pressure'] + dfcopy['Work Pressure']
```

```python
dfcopy['Satisfaction'] = dfcopy['Study Satisfaction'] + dfcopy['Job Satisfaction']
```

```python
dfcopy.drop(columns=['Academic Pressure','Work Pressure','Study Satisfaction','Job Satisfaction'], axis=1, inplace=True)
```

```python
dfcopy.head()
```

```python
dfcopy= dfcopy[dfcopy['Dietary Habits'].isin(['Moderate', 'Unhealthy', 'Healthy'])]
```

```python
dfcopy.drop(columns=['Sleep Duration'], axis=1, inplace=True)
```

```python
encoder = OrdinalEncoder()
```

```python
dfcopy.replace({'Gender':{'Female': 0,'Male': 1}}, inplace=True)
```

```python
dfcopy.replace({'Working Professional or Student':{'Working Professional': 0,'Student': 1},'Dietary Habits':{'Healthy':0,'Moderate':1,'Unhealthy':2}}, inplace=True)
```

```python
dfcopy.head()
```

```python
dfcopy.replace({'Family History of Mental Illness':{'No': 0,'Yes': 1},'Have you ever had suicidal thoughts ?':{'No': 0,'Yes': 1}}, inplace=True)
```

```python
dfcopy.head()
```

```python
dfcopy.info()
```

```python
X = dfcopy.drop(columns='Depression',axis=1)
X.shape
```

```python
y = dfcopy['Depression']
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)
```

```python
models = {
    'log_reg': LogisticRegression(max_iter=10000),
    'tree': DecisionTreeClassifier(),
    'RandomForestClassifier' : RandomForestClassifier(),
    'KNeighborsClassifier' :KNeighborsClassifier(),
    'svc': SVC(),
}
```

```python
for key, model in models.items():
    print(key)
    model.fit(X_train, y_train)
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    train_accuracy = accuracy_score(y_train, y_train_pred)
    test_accuracy = accuracy_score(y_test, y_test_pred)
    print(f"Train_accuracy: {train_accuracy}\n")
    print(f"Test_accuracy: {test_accuracy}\n")
    print("------------------------------------------")
    
    
```

```python
params = {
    'log_reg': {
        'penalty': ['l1', 'l2', 'elasticnet'],
        'C': [0.001, 0.01, 0.1, 1, 10, 100],
        'solver': ['liblinear', 'saga', 'lbfgs'],
        'max_iter': [100, 200, 500,1000],
        'l1_ratio': [0.1, 0.5, 0.7, 1.0]  # Add `l1_ratio` values to be used only with `elasticnet` penalty
    },
    'tree': {
        'criterion': ['gini', 'entropy'],
        'max_depth': [None, 10, 20, 30, 40, 50],
        'min_samples_split': [2, 5, 10, 20],
        'min_samples_leaf': [1, 2, 5, 10],
        'max_features': [None, 'sqrt', 'log2'],
        'max_leaf_nodes': [None, 10, 20, 50, 100],
    },
    'RandomForestClassifier': {
        'criterion': ['gini', 'entropy', 'log_loss'], 
        'max_depth': [None, 10, 20, 30, 40, 50],
        'min_samples_split': [2, 5, 10, 20],
        'min_samples_leaf': [1, 2, 5, 10],
        'max_features': [None, 'sqrt', 'log2'],
        'max_leaf_nodes': [None, 10, 20, 50, 100],
    }
}
```

```python
best_models = {}
```

```python
for model_name, model in models.items():
    print(f"Starting Grid Search for {model_name}...")
    rand_search= RandomizedSearchCV(
        estimator=model,
        param_distributions=params[model_name],
        scoring='accuracy',
        cv=5,
        n_jobs=-1 
    )
    
    # Fit GridSearchCV
    rand_search.fit(X_train, y_train)
    
    # Store the best model and score
    best_models[model_name] = {
        'best_params': rand_search.best_params_,
        'best_score': rand_search.best_score_,
        'best_estimator': rand_search.best_estimator_
    }
    
    print(f"Best parameters for {model_name}: {rand_search.best_params_}")
    print(f"Best score for {model_name}: {rand_search.best_score_}\n")
    print("------------------------------------------------------------------")
# Output the best parameters and scores for each model
print("Summary of Best Models:")
for model_name, info in best_models.items():
    print(f"{model_name}:")
    print(f"  Best Parameters: {info['best_params']}")
    print(f"  Best Score: {info['best_score']}")
    print("------------------------------------------------------------------")
```
