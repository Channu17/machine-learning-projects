```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train.csv')
dfcopy = df.copy()
```

```python
dfcopy.drop(columns=['id','Name', 'CGPA','Degree', 'Profession'], axis=1, inplace=True)
```

```python
dfcopy.drop(columns='City', axis=1, inplace= True)
```

```python
dfcopy['Financial Stress'].fillna(dfcopy['Financial Stress'].mean(),inplace=True)
```

```python
dfcopy['Study Satisfaction'].fillna(0,inplace=True)
dfcopy['Job Satisfaction'].fillna(0,inplace=True)
dfcopy['Academic Pressure'].fillna(0, inplace=True)
dfcopy['Work Pressure'].fillna(0, inplace=True)
```

```python
dfcopy['Pressure'] = dfcopy['Academic Pressure'] + dfcopy['Work Pressure']
```

```python
dfcopy['Satisfaction'] = dfcopy['Study Satisfaction'] + dfcopy['Job Satisfaction']
```

```python
dfcopy.drop(columns=['Academic Pressure','Work Pressure','Study Satisfaction','Job Satisfaction'], axis=1, inplace=True)
```

```python
def func(habit):
    if habit not in ['Moderate', 'Unhealthy', 'Healthy']:
        habit = 'Moderate'
    return habit 
dfcopy['Dietary Habits'] = dfcopy['Dietary Habits'].apply(func)
```

```python
def func2(sleep):
    if sleep not in ['Less than 5 hours', '7-8 hours', 'More than 8 hours','5-6 hours']:
        sleep = 'Less than 5 hours'
    return sleep
dfcopy['Sleep Duration'] = dfcopy['Sleep Duration'].apply(func2)
```

```python
dfcopy.head()
```

```python
dfcopy.replace({'Gender':{'Female': 0,'Male': 1}}, inplace=True)
```

```python
dfcopy.replace({'Working Professional or Student':{'Working Professional': 0,'Student': 1},'Dietary Habits':{'Healthy':0,'Moderate':1,'Unhealthy':2}}, inplace=True)
```

```python
dfcopy.replace({'Family History of Mental Illness':{'No': 0,'Yes': 1},'Have you ever had suicidal thoughts ?':{'No': 0,'Yes': 1}}, inplace=True)
```

```python
dfcopy.replace({'Sleep Duration':{'Less than 5 hours':4.5,'7-8 hours':7.5,'More than 8 hours':8.5,'5-6 hours':5.5}},inplace = True)
```

```python
dfcopy['Sleep Duration'].value_counts()
```

```python
X = dfcopy.drop(columns='Depression',axis=1)
X.shape
```

```python
y = dfcopy['Depression']
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2, stratify=y, random_state=42)
```

```python
import tensorflow as tf
import keras
from tensorflow.keras.callbacks import EarlyStopping
```

```python
from sklearn.preprocessing import StandardScaler

scalar = StandardScaler()
X_train_scaled = scalar.fit_transform(X_train)
X_test_scaled = scalar.transform(X_test)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape = (11,)),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(16, activation = 'relu'),
    tf.keras.layers.Dense(2, activation= 'sigmoid')
])
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
```

```python
model.fit(X_train_scaled, y_train,validation_data=(X_test_scaled, y_test), epochs=100, callbacks=early_stopping)
```

```python
loss, accuracy = model.evaluate(X_test_scaled, y_test)
print(accuracy)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten
from tensorflow.keras.optimizers import Adam

# Define a dictionary of hyperparameters
params = {
    'learning_rate': 0.001,           # Learning rate for the Adam optimizer
    'batch_size': 64,                 # Batch size for training
    'epochs': 20,                     # Number of epochs for training
    'dropout_rate': 0.3,              # Dropout rate for regularization
    'hidden_layers': [128, 64],       # Number of neurons in hidden layers
    'activation': 'relu',             # Activation function for hidden layers
    'output_activation': 'softmax',   # Activation function for output layer
    'optimizer': 'adam',              # Optimizer to use ('adam', 'sgd', etc.)
    'loss_function': 'sparse_categorical_crossentropy',  # Loss function for classification
    'metrics': ['accuracy'],         # Metrics to monitor during training
    'early_stopping': True,           # Whether to use early stopping
    'patience': 3                     # Patience for early stopping
}

# Build the ANN model
def build_model(params):
    model = Sequential()

    # Input layer (Flatten the input, assuming input shape is (28, 28) for MNIST)
    model.add(Flatten(input_shape=(11,)))

    # Add hidden layers
    for units in params['hidden_layers']:
        model.add(Dense(units, activation=params['activation']))
        model.add(Dropout(params['dropout_rate']))  # Dropout for regularization

    # Output layer
    model.add(Dense(10, activation=params['output_activation']))  # Assume 10 classes for classification

    # Compile the model with chosen optimizer and loss function
    if params['optimizer'] == 'adam':
        optimizer = Adam(learning_rate=params['learning_rate'])
    else:
        # For SGD or other optimizers, you can modify this as needed
        optimizer = tf.keras.optimizers.SGD(learning_rate=params['learning_rate'])

    model.compile(optimizer=optimizer,
                  loss=params['loss_function'],
                  metrics=params['metrics'])

    return model

# Create the model
model = build_model(params)

# Print the model summary to check architecture
model.summary()

# If using early stopping, define it
if params['early_stopping']:
    early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=params['patience'], restore_best_weights=True)
else:
    early_stopping = None

# Train the model
history = model.fit(
    # X_train and y_train should be your training data and labels
    X_train, y_train, 
    batch_size=params['batch_size'],
    epochs=params['epochs'],
    validation_data=(X_test, y_test),  # X_val and y_val should be your validation data
    callbacks=[early_stopping] if early_stopping else None
)

# Evaluate the model on test data
# X_test and y_test should be your test data and labels
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f"Test accuracy: {test_acc}")
```
